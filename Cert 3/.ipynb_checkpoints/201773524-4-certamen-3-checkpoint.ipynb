# Pregunta 1 (valores en Aula con 5 decimales aproximados)

Resolviendo la siguiente ecuación en $x \in ]-1,1[$:

$$y''(x) = (y(x) - 1)(1 + (y'(x))^2)^{3/2}$$

se obtiene el valor de $w_{2dot} = (w_1 - 1)(1+w_2^2)^\frac{3}{2}$

```python
#Se usa el método del disparo y se modifican las funciones del git del curso para contemplar el nuevo sistema
#abordado en el rango [-1,1]



# RHS of dynamical system
def my_f1(t,w):
    w1 = w[0]
    w2 = w[1]
    w1dot = w2
    w2dot = (w1-1)*((1+(np.power(w2,2)))**(3*(1/2)))
    return np.array([w1dot,w2dot])

# Function to be used to apply the Shooting Method
def F_SM_1(alpha,y0,y1,N):
    t = np.linspace(-1,1,N)
    initial_condition = np.zeros(2)
    initial_condition[0] = y0
    initial_condition[1] = alpha
    sol = solve_ivp(my_f1,(-1,1),initial_condition, t_eval = t)
    return sol.y[0,-1]-y1



'''
input:
N        : (int) número de puntos de discretización del intervalo [-1,1]

output:
x        : (ndarray) discretización del intervalo espacial [-1,1] en N puntos
y        : (ndarray) aproximación numérica de y(x) en los N puntos en que se discretizó 
            el intervalo espacial [-1,1]
'''
def solve_2nd_order_ODE(N):
    # Your own code.
    #se modifica el valor arbitrario en root a 1
    F_root_1 = lambda alpha: F_SM_1(alpha,0,0,N)
    alpha_r = root(F_root_1, 1).x[0]

    sol = solve_ivp(my_f1,(-1,1),np.array([0,alpha_r]), t_eval = np.linspace(-1,1,N))
    x = sol.t
    y = sol.y[0,:]
    return x, y

'''
input:
N        : (int) número de puntos de discretización del intervalo [-1,1]

output:
integral : (double) aproximación numérica de la integral de y(x) entre -1 y 1.
'''
def compute_integral(N):
    # Your own code.
    # se usa la funcion trapezoid para ser evaluada en los resultados de la primera función implementada
    x,y = solve_2nd_order_ODE(N)
    integral = trapezoid(y,x)
    return integral



'''
input:
N        : (int) número de puntos de discretización del intervalo [-1,1]

output:
max_y    : (double) máximo valor de y(x) obtenido en la discretización espacial de N puntos de intervalo [-1,1],
            si existieran 2 máximos solo retornar el primero.
max_x    : (double) coordenada x donde se alcanza el máximo valor de y(x)
'''
def find_max(N):
    # Your own code.
    #se busca el mayor valor en y, luego el indice de ese valor se obtiene la coordenada x
    x,y = solve_2nd_order_ODE(N)
    max_y = 0.0
    i = 0
    for valor in y:
        if(valor > max_y):
            max_y = valor
            i += 1
    max_x = x[i]
    return max_y, max_x


'''
input:
m       : (double) pendiente objetivo
N       : (double) número de puntos de discretización del dominio [-1,1]
delta   : (double) umbral considerado para aceptar la aproximación de la pendiente m, 
            es decir |y(widehat_x)-m|<delta

output:
widehat_x      : (double) solución de la ecuación y(widehat_x)=m, considere -1<widehat_x<1
'''
def find_widehat_x(m, N, delta):
    # Your own code.
    #se usara central difference y un flag en caso de que no se encuentre una diferencia menor a delta
    x,y = solve_2nd_order_ODE(N)
    flag = False
    for valor in range(1,N-1):
        h = abs(x[2]-x[1])
        central = (y[valor+1] - y[valor-1])/(2*h)
        d = abs(central - m)
        if d < delta:
            flag = True
            widehat_x = x[valor]
            break
    if flag == False:
        widehat_x = np.nan    
    return widehat_x
```

# Pregunta 2

Resolviendo el sistema nos queda:

$-Q*A^{-1} * X^{m+1} + X^{m+1} * A^{-1} * X^M = -A^T$

$A_2 * X^{m+1} + X^{m+1} * B = C$


```python
#funciones requeridas obtenidas del git del curso (bonus 9) (sylvester equation using GMRes)
def inverse_matrix(A):
    Q,R=qr(A) 
    return sp.linalg.solve_triangular(R,Q.T)
def compute_matrix_vector_product(x,A,B,Q):
    X = np.reshape(x,(n,n),order='F')
    out = np.dot(A,X)+np.dot(X,B)
    return out.flatten('F')
'''
input:
A       : (ndarray) matriz input A
Q       : (ndarray) matriz input Q
Xm      : (ndarray) matriz input X^{(m)}
n       : (int) se considera que las dimensiones de las matrices son 'n x n'

output:
Xout         : (ndarray) matriz de n x n output X^{(m+1)}
rel_residual : (double)  residuo relativo del sistema de ecuaciones lineales.
'''
def obtain_next_X(A,Q,Xm,n):
    # Your own code.
    #se aplica QR para encontrar la inversa de A
    A_1 = inverse_matrix(A)
    A_2 = np.dot(-Q,A_1)
    B = np.dot(A_1,Xm)
    C = -np.transpose(A)
    Ax = lambda x: compute_matrix_vector_product(x,A_2,B,n)
    # This is the famous 'afun'!!
    afun = LinearOperator((n**2, n**2), matvec=Ax)
    # Just running GMRes
    x, exitCode = gmres(afun, C.flatten('F'), tol=1e-10)
    # Just reshaping the 
    X_GMRes = np.reshape(x,(n,n),order='F')
    res = np.linalg.norm(np.dot(A_2,X_GMRes) + np.dot(X_GMRes,B)-C)
    rel_residual = res
    Xout = X_GMRes
    return Xout, rel_residual

```
